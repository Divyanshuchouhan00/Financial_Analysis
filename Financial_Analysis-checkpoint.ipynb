```python
import matplotlib.pyplot as plt  # Bar chart for market cap comparison:

df.groupby('Market_Cap_Category')['Mar Cap - Crore'].mean().plot(kind='bar', color='skyblue')
plt.title("Average Market Capitalization by Category")
plt.ylabel("Market Cap (Crores)")
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby('Sales_Qrt_Category')['Sales Qtr - Crore'].mean().plot(kind='bar', color='lightgreen')
plt.title("Average Quarterly Sales by Category")
plt.ylabel("Sales (Crores)")
plt.xticks(rotation=0)
plt.show()      #Bar chart for quarterly sales comparison:
```

```python
import matplotlib.pyplot as plt

# Define color mapping for categories
colors = {'Large Cap': 'blue', 'Mid Cap': 'green', 'Small Cap': 'red'}

# Create scatter plot with two colors based on Market_Cap_Category
for category, color in colors.items():
    subset = df[df['Market_Cap_Category'] == category]
    plt.scatter(subset['Mar Cap - Crore'], subset['Sales Qtr - Crore'], label=category, color=color, alpha=0.6)

# Labels & Titles
plt.xlabel("Market Capitalization (Crore)")
plt.ylabel("Quarterly Sales (Crore)")
plt.title("Market Cap vs Quarterly Sales (Segmented by Category)")
plt.legend()  # Show legend
plt.grid()

# Show plot
plt.show()  #Helps compare company performance visually across market categories
```
